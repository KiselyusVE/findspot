Подключим необходимые библиотеки

```python
import requests
import os
import pandas as pd
```

Сформируем список интересующих нас дат

```python
daterange = pd.date_range('2011/01/01', '2016/01/01', freq = 'D')
print(str(daterange[0]).split('-'))
```

Создадим необходимые для сохранения директории

```python
path_to_save = './imgs_full/'
```

```python
for year in range(2011, 2016):
    if not os.path.exists(path_to_save + str(year)):
        os.makedirs(path_to_save + str(year))
```

В этом цикле мы проходим по датам и скачиваем снимки Солнца в папки, названия которых соответсвуют годам

```python
for date in daterange:

    year = str(date)[:-9].split('-')[0]
    month = str(date)[:-9].split('-')[1]
    day = str(date)[:-9].split('-')[2]

    url = 'http://suntoday.lmsal.com/sdomedia/SunInTime/' + year + '/' + month + '/' + day + '/f_HMI_cont_aiascale.jpg'
    filename = year + '_' + month + '_' + day + '.jpg'

    r = requests.get(url, allow_redirects=True)
    open(path_to_save + year + '/' + filename, 'wb').write(r.content)
    print(filename + '+')
    
```
